```python
pin = int(input('silahkan masukkan pin: '))
while pin != 1234:
  print('pin salah')
  pin = int(input('silahkan masukkan pin: '))
if pin == 1234:
  print('pin sudah benar')
```

```python
angka = int(input('masukkan angka:'))
coba = 1
print('percobaan ke-', coba)
while angka != 6 and coba <5:
  print()
  angka = int(input('angka salah, coba lagi: '))
  coba+= 1
  print('percobaan ke-', coba)

if angka == 6:
  print('benar')
else:
  print('gagal')
```

99 botol

```python
for i in range(99,0,-1):
  print(f'{i} botol di tembok')
  print(f'{i} botol')
  print('ambil 1 botol')
  print(f'{i-1} botol di tembok')
  print('')

a = 99
while a>0:
  print(f'{a} botol di tembok')
  print(f'{a} botol')
  print('ambil 1 botol')
  print(f'{a-1} botol di tembok')
  a -= 1
  print('')
```

fizzbuzz

```python
for i in range(1,101):
  if i % 3 == 0 and i % 5 ==0:
    print('fizzbuzz')
  elif i % 3 == 0:
    print('fizz')
  elif i % 5 == 0:
    print('buzz')
  else:
    print(i)

print('')

a = 1
while a<=100:
  if a%3==0 and a%5==0:
    print('fizzbuzz')
  elif a%3==0:
    print('fizz')
  elif a%5==0:
    print('buzz')
  else:
    print(a)
  a += 1
```
