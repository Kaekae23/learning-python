```python
class restoran:
  def __init__(self,nama):
    self.nama = nama
    self.menu = []
  def buka(self):
    print(f"{self.nama} telah buka")
  def makan(self,makanan,harga):
    self.menu.append((makanan,harga))
    print(f"{self.nama} telah menyediakan menu {makanan} dengan harga {harga:,}")
  def daftar(self):
    print(f"menu di {self.nama}")
    for i in self.menu:
      print(f"- {i[0]:<15} Rp.{i[1]:,}")

a = restoran("bakso mang ujang")
a.buka()
a.makan("bakso jumbo",15000)
a.makan("mi ayam bakso", 18000)
a.makan("bakso beranak",25000)
a.daftar()
```

```python
class resto:
  def __init__(self,nama):
    self.nama = nama
    self.menu = []
  def buka(self):
    print(f"warung {self.nama} telah buka")
  def daftar_menu(self,makanan,harga):
    self.menu.append((makanan,harga))
    print(f"menu yang ditambahkan {makanan}, harga {harga}")
  def lihat_menu(self):
    print("daftar menu".center(30))
    for i in self.menu:
      print(f'- {i[0]:<15}Rp. {i[1]}')
  def pesan(self,pesanan):
    daftar = []
    for i in self.menu:
      daftar.append(i[0])
    if pesanan in daftar:
      print("pesanan disiapkan")
    else:
      print("maaf pesanan tidak tersedia")


a= resto("bakso sabar")
a.buka()
a.daftar_menu("nasi goreng",18000)
a.daftar_menu("bakso beranak", 20000)
a.daftar_menu("mi ayam", 17000)
a.daftar_menu("bihun goreng",12000)
a.lihat_menu()
a.pesan("mi ayam")
a.pesan("bebek")
```

```python
class resto:
  def __init__(self,nama,lokasi):
    self.nama = nama
    self.lokasi = lokasi
    self.menu = []
  def buka(self):
    print(f'warung {self.nama} buka di {self.lokasi}')
  def makan(self,makanan,harga):
    self.menu.append((makanan,harga))
  def lihat_menu(self):
    print("Daftar Menu".center(30))
    for i in self.menu:
      print(f'{i[0]:<15} Rp.{i[1]:,}')

class restoran(resto):
  def __init__(self,nama,lokasi,jam):
    super().__init__(nama,lokasi)
    self.jam = jam
  def sambutan(self):
    print(f'selamat datang di warung {self.nama}, buka mulai pukul {self.jam}')

a= restoran("mang ujang","kebayoran lama",9)
a.sambutan()
a.buka()
a.makan("bakso",12000)
a.makan("bihun goreng",10000)
a.makan("mi ayam",20000)
a.lihat_menu()
```

```python
class country:
  def __init__(self,negara):
    self.negara = negara
  def tampilkan_negara(self):
    print(f'nama negara {self.negara}')
class city(country):
  def __init__(self,negara,kota,alasan):
    super().__init__(negara)
    self.kota = kota
    self.alasan = alasan
  def tampilkan_kota(self):
    print(f"kota : {self.kota}, alasan : {self.alasan}")
class kota_favorit:
  def __init__(self):
    self.daftar_kota = []
  def tambah_kota(self,tempat):
    self.daftar_kota.append(tempat)
  def lihat(self):
    print("daftar tempat favorit".center(50))
    for i in self.daftar_kota:
      print(f"{i.kota:<15},{i.negara:<10}->{i.alasan:>20}")

a = city("indo","bandung","dingin dan sejuk")
b = city("malay","kl","bagus dan bersih")
c = city("brunei","bru","nyaman dan damai")
a1 = kota_favorit()
a1.tambah_kota(a)
a1.tambah_kota(b)
a1.tambah_kota(c)
a1.lihat()


```

bank account

```python
class bankaccount:
  def __init__(self,pemilik):
    self.pemilik = pemilik
    self.saldo = 0
  def deposit(self,jumlah):
    self.saldo += jumlah
  def withdraw(self,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
    else:
      print(f'maaf, saldo tidak mencukupi. Saldo saat ini {self.saldo}')
  def tampilkan_saldo(self):
    print(f'saldo anda saat ini {self.saldo}')
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      penerima.deposit(jumlah)
      print(f'transfer berfasil, saldo anda saat ini {self.saldo}')
    else:
      print(f'maaf, saldo tidak mencukupi')

orang1 = bankaccount("ana")
orang2 = bankaccount("doni")
orang1.deposit(50000)
orang2.deposit(200000)
orang1.tampilkan_saldo()
orang2.withdraw(10000)
orang2.tampilkan_saldo()
orang2.transfer(orang1,20000)
orang1.tampilkan_saldo()
```

akun bank (tabungan & bisnis)

```python
class bankaccount:
  def __init__(self,pemilik):
    self.pemilik = pemilik
    self.saldo = 0
  def deposit(self,jumlah):
    self.saldo += jumlah
    print(f'saldo anda saat ini Rp. {self.saldo:,}')
  def withdraw(self,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
    else:
      print(f'maaf, saldo anda tidak mencukupi untuk penarikan')
  def lihat_saldo(self):
    print(f'saldo anda saat ini {self.saldo}')
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      penerima.deposit(jumlah)
      print(f'anda mengirim uang ke: {penerima.pemilik} saldo anda saat ini {self.saldo}')
    else:
      print(f'maaf, saldo anda tidak mencukupi untuk melakukan transfer')

class bisnis(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)
    self.biaya_admin = 5000
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah + self.biaya_admin:
      self.saldo -= jumlah + self.biaya_admin
      print(f'anda telah melakukan transfer ke: {penerima.pemilik} sebesar Rp. {jumlah:,} dan biaya admin sebesar Rp. {self.biaya_admin:,}')
      print(f'saldo anda saat ini {self.saldo}')
      penerima.deposit(jumlah)
    else:
      print(f'maaf, saldo anda tidak mencukupi')

class tabungan(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)

orang1 = bisnis("nana")
orang2 = tabungan('danu')
orang1.deposit(400000)
orang2.deposit(230000)
orang1.transfer(orang2,35000)



```

```python
class bankaccount():
  def __init__(self,pemilik):
    self.pemilik = pemilik
    self.riwayat = []
    self.saldo = 0
  def deposit(self,jumlah):
    self.saldo += jumlah
    self.riwayat.append(f'deposit : Rp. {jumlah:,}')
  def withdraw(self,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      self.riwayat.append(f'tarik : Rp {jumlah:,}')
    else:
      print(f'saldo tidak cukup')
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      self.riwayat.append(f'transfer ke {penerima.pemilik} : Rp. {jumlah:,}')
      penerima.saldo += jumlah
      penerima.riwayat.append(f'menerima transfer dari {self.pemilik} : Rp. {jumlah:,}')
    else:
      print(f'saldo tidak cukup')
  def tampilkan_saldo(self):
    print(f'saldo anda {self.pemilik} saat ini : Rp. {self.saldo:,}')
  def tampilkan_riwayat(self):
    print(f'riwayat transaksi {self.pemilik}:')
    for i in self.riwayat:
      print(f'- {i}')

class bisnis(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)
    self.biaya_admin = 5000
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah + self.biaya_admin:
      self.saldo -= jumlah + self.biaya_admin
      self.riwayat.append(f'transfer ke {penerima.pemilik} : Rp. {jumlah:,}, biaya admin Rp. {self.biaya_admin:,}')
      penerima.saldo += jumlah
      penerima.riwayat.append(f'menerima transfer dari {self.pemilik} : Rp. {jumlah:,}')
    else:
      print(f'saldo tidak cukup')

class tabungan(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)

orang1 = bisnis("danu")
orang2 = tabungan("nina")
orang3 = tabungan("doni")
orang4 = bisnis("radi")
orang1.deposit(500000)
orang2.deposit(750000)
orang1.transfer(orang3,200000)
orang3.deposit(710000)
orang3.transfer(orang4,25000)
orang4.withdraw(160000)
orang1.tampilkan_saldo()
orang1.tampilkan_riwayat()
orang2.tampilkan_saldo()
orang2.tampilkan_riwayat()
orang3.tampilkan_saldo()
orang3.tampilkan_riwayat()
orang4.tampilkan_saldo()
orang4.tampilkan_riwayat()
```

```python
class bankaccount():
  def __init__(self,pemilik):
    self.pemilik = pemilik
    self.riwayat = []
    self.saldo = 0
    self.denda = 10000
  def deposit(self,jumlah):
    self.saldo += jumlah
    self.riwayat.append(f'deposit : Rp. {jumlah:,}')
  def withdraw(self,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      self.riwayat.append(f'tarik : Rp {jumlah:,}')
    else:
      print(f'saldo tidak cukup')
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah:
      self.saldo -= jumlah
      self.riwayat.append(f'transfer ke {penerima.pemilik} : Rp. {jumlah:,}')
      penerima.saldo += jumlah
      penerima.riwayat.append(f'menerima transfer dari {self.pemilik} : Rp. {jumlah:,}')
    else:
      print(f'saldo tidak cukup')
  def tampilkan_saldo(self):
    print(f'saldo anda {self.pemilik} saat ini : Rp. {self.saldo:,}')
  def tampilkan_riwayat(self):
    print(f'riwayat transaksi {self.pemilik}:')
    for i in self.riwayat:
      print(f'- {i}')
  def tambah_bunga(self):
    bunga = self.saldo * 0.05
    self.saldo += bunga
    self.riwayat.append(f'bunga 5% : Rp. {bunga:,.0f}')

class bisnis(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)
    self.biaya_admin = 5000
  def transfer(self,penerima,jumlah):
    if self.saldo >= jumlah + self.biaya_admin:
      self.saldo -= jumlah + self.biaya_admin
      self.riwayat.append(f'transfer ke {penerima.pemilik} : Rp. {jumlah:,}, biaya admin Rp. {self.biaya_admin:,}')
      penerima.saldo += jumlah
      penerima.riwayat.append(f'menerima transfer dari {self.pemilik} : Rp. {jumlah:,}')
    else:
      print(f'saldo tidak cukup')
  def tambah_denda(self):
    if self.saldo < 100000:
      self.saldo -= self.denda
      self.riwayat.append(f'denda : Rp. {self.denda:,} ')

class tabungan(bankaccount):
  def __init__(self,pemilik):
    super().__init__(pemilik)

orang1 = bisnis("danu")
orang2 = tabungan("nina")
orang3 = tabungan("doni")
orang4 = bisnis("radi")
orang1.deposit(500000)
orang2.deposit(750000)
orang1.transfer(orang3,200000)
orang3.deposit(710000)
orang3.transfer(orang4,25000)
orang4.withdraw(160000)
orang1.tambah_bunga()
orang1.tampilkan_saldo()
orang1.tampilkan_riwayat()
orang2.tampilkan_saldo()
orang2.tampilkan_riwayat()
orang3.tampilkan_saldo()
orang3.tampilkan_riwayat()
orang4.tambah_denda()
orang4.tampilkan_saldo()
orang4.tampilkan_riwayat()

```

pokedex

```python
class pokemon:
  def __init__(self,nama,tipe,hp,attack,defense):
    self.nama = nama
    self.tipe = tipe
    self.hp = hp
    self.attack = attack
    self.defense = defense
  def info(self):
    print(f'''
    nama : {self.nama}
    tipe : {self.tipe}
    hp : {self.hp}
    attack : {self.attack}
    defense : {self.defense}
    ''')
  def serang(self,target):
    damage = self.attack- target.defense
    if damage <= 0:
      damage = 0
    target.hp -= damage
    if target.hp <= 0:
      target.hp = 0
    print(f'{self.nama} menyerang {target.nama} dengan damage {damage}')
    print(f'hp {target.nama} tersisa {target.hp}')

class pokedex:
  def __init__(self):
    self.daftar_pokemon = []
  def tambah_pokemon(self,pokemon):
    self.daftar_pokemon.append(pokemon)
  def tampilkan_semua(self):
    for i in self.daftar_pokemon:
      i.info()
  def cari_pokemon(self,nama):
    nama_pokemon = []
    for i in self.daftar_pokemon:
      nama_pokemon.append(i.nama.lower())
    if nama.lower() in nama_pokemon:
      for i in self.daftar_pokemon:
        if nama.lower() == i.nama.lower():
          i.info()
    else:
      print(f'{nama} tidak ada dalam daftar')


a1 = pokemon('pikachu','petir',1000,120,97)
a2 = pokemon('mabsposa','tanah',970,100,89)
a3 = pokemon('trasso','air',1020,210,150)
pokedex = pokedex()
pokedex.tambah_pokemon(a1)
pokedex.tambah_pokemon(a2)
pokedex.tambah_pokemon(a3)
a1.serang(a2)
pokedex.cari_pokemon('mabsposa')
pokedex.tampilkan_semua()
pokedex.cari_pokemon('kala')
```
