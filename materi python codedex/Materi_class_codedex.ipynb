restoran

```python
class resto:
  nama =''
  menu = ''
  rating = 0.0
  online = False

kebab_ahmad = resto()
kebab_ahmad.nama = 'kebab ahmad'
kebab_ahmad.menu = 'kebab'
kebab_ahmad.rating = 4.9
kebab_ahmad.online = True

bakmi_ujang = resto()
bakmi_ujang.nama = 'bakmi abang ujang'
bakmi_ujang.menu = 'bakmi'
bakmi_ujang.rating = 4.86
bakmi_ujang.online = True

print(vars(kebab_ahmad))
print(vars(bakmi_ujang))
print()

for i in vars(kebab_ahmad):
  print(i)
  print(vars(kebab_ahmad)[i])
  print()
print()

for i in vars(bakmi_ujang):
  print(f'{i:<10} : {vars(bakmi_ujang)[i]}')
```

```python
class akun:
  def __init__(self,nama,pin,saldo):
    self.nama = nama
    self.pin = pin
    self.saldo = saldo

  def setor(self,jumlah):
    self.saldo += jumlah

  def tarik(self,jumlah):
    self.saldo -= jumlah

  def transfer(self,jumlah,penerima):
    self.saldo -= jumlah
    penerima.setor(jumlah)

  def cek(self):
    print(f'saldo saat ini : Rp. {self.saldo:,}')

akun1 = akun('nana',777633,50000)
akun2 = akun('neni',434343,10250000)
akun1.transfer(200000,akun2)
akun1.setor(2007000)
akun1.cek()
akun1.tarik(200000)
akun1.cek()
akun2.cek()
```

```python
class akun:
  def __init__(self,nama,pin,saldo):
    self.nama = nama
    self.pin = pin
    self.saldo = saldo
    self.riwayat_transaksi = [f'awal : {saldo:,}']
    self.riwayat_saldo = [saldo]

  def setor(self,jumlah):
    self.saldo += jumlah
    self.riwayat_transaksi.append(f'setor : +{jumlah:,}')
    self.riwayat_saldo.append(self.saldo)

  def tarik(self,jumlah):
    self.saldo -= jumlah
    self.riwayat_transaksi.append(f'tarik : -{jumlah:,}')
    self.riwayat_saldo.append(self.saldo)

  def transfer(self,penerima,jumlah):
    self.saldo -= jumlah
    penerima.setor(jumlah)
    self.riwayat_transaksi.append(f'transfer : -{jumlah:,}')
    self.riwayat_saldo.append(self.saldo)

  def cek_saldo(self):
    print(f'{self.saldo:,}')

  def riwayat(self):
    print(f"{'status':<6} {'transaksi':<10} {'saldo':<10}")
    print('-'*30)
    for i in range(len(self.riwayat_saldo)):
      print(f'{self.riwayat_transaksi[i]:<10}  {self.riwayat_saldo[i]:,} ')

akun1 = akun('riri',2232,100000)
akun2 = akun('ruru',2424,23000)
akun1.setor(100000)
akun1.tarik(50000)
akun1.transfer(akun2,70000)
akun1.cek_saldo()
akun1.riwayat()
akun2.riwayat()
```

```python
class rekening:
  def __init__(self,nama,saldo,pin):
    self.nama = nama
    self.saldo = saldo
    self.pin = pin
    self.riwayat_saldo = [saldo]
    self.riwayat_transaksi = [f"{'awal':<8} : {saldo:,}"]

  def setor(self,jumlah):
    self.saldo += jumlah
    self.riwayat_saldo.append(self.saldo)
    self.riwayat_transaksi.append(f"{'setor':<8} : {jumlah:,}")

  def tarik(self,jumlah):
    self.saldo -= jumlah
    self.riwayat_saldo.append(self.saldo)
    self.riwayat_transaksi.append(f"{'tarik':<8} : {jumlah:,}")

  def transfer(self,jumlah,penerima):
    self.saldo -= jumlah
    penerima.setor(jumlah)
    self.riwayat_saldo.append(self.saldo)
    self.riwayat_transaksi.append(f"{'transfer':<8} : {jumlah:,} ke {penerima.nama}")

  def cek_saldo(self):
    print(f'saldo saat ini : {self.saldo}')

  def cek_riwayat(self):
    print(f'riwayat transaksi {self.nama}')
    print('-'*30)
    for i in range(len(self.riwayat_saldo)):
      print(f'{self.riwayat_transaksi[i]} --- {self.riwayat_saldo[i]}')

akun1 = rekening('lulu',100000,1222)
akun2 = rekening('dodo',200000,5656)
akun1.setor(480000)
akun2.tarik(50000)
akun1.transfer(340000,akun2)
akun1.cek_saldo()
akun2.cek_saldo()
akun1.cek_riwayat()

akun2.cek_riwayat()
```

pokedex

```python
class pokemon:
  def __init__(self,entry,nama,tipe,deskripsi,tertangkap):
    self.entry = entry
    self.nama = nama
    self.tipe = tipe
    self.deskripsi = deskripsi
    self.tertangkap = tertangkap

  def ucap_nama(self):
    print(f'{self.nama}, {self.nama}!')

  def tampil(self):
    print(f'entry : {self.entry}')
    print(f'nama: {self.nama}')
    tampil_tipe = ', '.join(self.tipe)
    print(f'tipe: {tampil_tipe}')
    print(f'deskripsi : {self.deskripsi}')
    if self.tertangkap:
      print(f'status : Tertangkap')
    else:
      print(f'status : Belum Tertangkap')

pikachu = pokemon(12,'pikachu',['api','air','listrik'],'penghantar listrik',False)
pikachu.ucap_nama()
pikachu.tampil()
```

```python
class pokemon:
  def __init__(self,nama,tipe,deskripsi,tertangkap):
    self.nama = nama
    self.tipe = tipe
    self.deskripsi = deskripsi
    self.tertangkap = tertangkap

  def panggil_nama(self):
    print(f'{self.nama}, {self.nama}')

  def tampil(self):
    print(f'nama : {self.nama}')
    print(f"tipe : {', '.join(self.tipe)}")
    print(f'deskripsi : {self.deskripsi}')
    if self.tertangkap:
      print(f'status : tertangkap')
    else:
      print(f'status : belum tertangkap')

pikachu = pokemon('pikachu',['tanah','api'],'jaga hati jaga pikiran',True)
pikachu.tampil()
```
