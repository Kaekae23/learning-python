```python
import random

pilihan = ['🍎','🍍','🍄','🌼']
hasil = random.choices(pilihan, k=3)
print(' - '.join(hasil))
if hasil[0] == hasil[1] == hasil[2]:
  print('selamat')
else:
  print('coba lagi')
```

mesin acak + while

```python
import random
def main():
  for i in range(50):
    pilihan = ['🍎','🍍','🍄','🌼']
    hasil = random.choices(pilihan, k=3)
    print(' - '.join(hasil))
    if hasil[0] == hasil[1] == hasil[2]:
      print(f'selamat anda berhasil di percobaan ke-{i+1}')
      break

user = ''
while user.lower() != 'n':
  main()
  user = input('mau main lagi (y/n)? ')
```

solar system

```python
from math import pi
from random import choice as ch

planet = ['merkurius','venus','bumi','mars','jupiter']

acak = ch(planet)

if acak =='merkurius':
  radius = 2440
elif acak == 'venus':
  radius = 6052
elif acak == 'bumi':
  radius = 6371
elif acak == 'mars':
  radius = 3390
elif acak == 'jupiter':
  radius = 28962
print(f'{pi:.2f}')
luas_planet = 4* pi * (radius**2)
print(f'luas planet {acak} : {luas_planet:,.2f} km kuadrat')
```

```python
import math
import random
#cara 1
planet = ['merkurius','venus','bumi','mars','jupiter']
radius = [2440,6052,631,3390,28962]
a= random.randint(0,len(planet)-1)
luas = math.pi * (radius[a]**2)
print(f'luas planet {planet[a]} dengan jari jari {radius[a]} km: {luas:,.2f} km')

#cara2
a = random.choice(planet)

if a =='merkurius':
  radius = 2440
elif a == 'venus':
  radius = 6052
elif a == 'bumi':
  radius = 631
elif a == 'mars':
  radius = 3390
elif a == 'jupiter':
  radius = 28962
luas = (radius**2) * math.pi
print(f'luas planet {a} dengan jari jari {radius} km : {luas:,.2f} km')
```

datetime

```python
import datetime

today = datetime.date.today()
ultah = datetime.date(2026,4,11)
lahir = datetime.date(2003,4,11)
usia = today - lahir
usia_tahun = usia.days // 365
usia_bulan = (usia.days % 365) // 30
usia_hari = (usia.days % 365) % 30
waktu = ultah - today
print(f'waktu tersisa sampai nambah umur : {waktu.days} hari')
print(f'usia saat ini :  {usia_tahun} tahun {usia_bulan} bulan {usia_hari} hari')
```
