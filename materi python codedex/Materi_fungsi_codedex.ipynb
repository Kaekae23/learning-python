```python
def tambah(a,b):
  return a+b

def kurang(a,b):
  return a-b

def kali(a,b):
  return a*b

def bagi(a,b):
  return a/b

def pangkat(a,b):
  return a**b

pertambahan = tambah(5,9)
pengurangan = kurang(34,5)
perkalian = kali(383,22)
pembagian = bagi(588,54)
perpangkatan = pangkat(55,4)

print(f'hasil pertambahan = {pertambahan}')
print(f'hasil pengurangan = {pengurangan}')
print(f'hasil perkalian = {perkalian:,}')
print(f'hasil pembagian = {pembagian:.2f}')
print(f'hasil perpangkatan = {perpangkatan:,}')
```

```python
daftar = [12,34,11,83,8,191,99,199]

def list(i):
  return daftar[i-1]

def minimal(a,b):
  mn = daftar[0]
  mx = daftar[0]
  for i in range(a,b+1):
    mn = min(mn,list(i))
    mx = max(mx,list(i))
  return mn,mx

terkecil,terbesar = minimal(1,7)
print(terkecil)
print(terbesar)
```

```python
def menu():
  print(f'''
selamat datang
silahkan memesan
1. bakso
2. nasi goreng
3. mi ayam
4. kwetiau goreng
5. kwetiau rebus''')


user = 'bakso 2, kwetiau goreng 3, mi ayam 1'
pesanan = user.split(', ')
jumlah = []
makanan = []
for pesan in pesanan:
  a = [i for i in pesan]
  jumlah.append(int(a[-1]))
  a.pop()
  if a[-1] == ' ':
    a.pop(-1)
  a = ''.join(a)
  makanan.append(a)
print(jumlah)
print(makanan)

def daftar_harga(makanan,jumlah):
  harga = []
  total_bayar = []
  for i in range(len(makanan)):
    if makanan[i] == 'bakso':
      harga.append(15000)
      bayar = 15000*jumlah[i]
      total_bayar.append(bayar)
    elif makanan[i] == 'nasi goreng':
      harga.append(13000)
      bayar = 13000*jumlah[i]
      total_bayar.append(bayar)
    elif makanan[i] == 'mi ayam':
      harga.append(16000)
      bayar = 16000*jumlah[i]
      total_bayar.append(bayar)
    elif makanan[i] == 'kwetiau goreng':
      harga.append(20000)
      bayar = 20000*jumlah[i]
      total_bayar.append(bayar)
    elif makanan[i] == 'kwetiau rebus':
      harga.append(25000)
      bayar = 25000*jumlah[i]
      total_bayar.append(bayar)
  return harga,total_bayar

def tampil(makanan,jumlah,harga,total_bayar):
  for i in range(len(makanan)):
    print(f'{makanan[i]:<20} : {harga[i]:,} x {jumlah[i]:<3} = {total_bayar[i]:,}')

harga, total_bayar = daftar_harga(makanan,jumlah)
tampil(makanan,jumlah,harga,total_bayar)
print('-'*45)
print(f'{'total':<33} {sum(total_bayar):>10}')
```

```python
menu = ['bakso','nasi goreng','mi ayam','kwetiau goreng','kwetiau rebus']
harga = [15000,13000,16000,20000,25000]
user = input('masukkan pesanan : ').lower()
pesanan = user.split(', ')
jumlah = []
makanan = []
for pesan in pesanan:
  a = [i for i in pesan]
  jumlah.append(int(a[-1]))
  a.pop()
  if a[-1] == ' ':
    a.pop(-1)
  makanan.append(''.join(a))

def total_bayar(jumlah,makanan,menu,harga):
  bayar = []
  total = []
  item = 0
  for i in range (len(makanan)):
    for j in range (len(menu)):
      if makanan[i] == menu[j]:
        item = harga[j]
        bayar.append(item)
        total.append(item*jumlah[i])
  return bayar,total

def pembayaran(bayar,total,makanan,jumlah):
  for k in range(len(makanan)):
    print(f'{makanan[k]:<15} : {bayar[k]:,} x {jumlah[k]} = {total[k]:,}')
  print('-'*45)
  print(f'{"total":<30} {sum(total):,}')


bayar,total = total_bayar(jumlah,makanan,menu,harga)
pembayaran(bayar,total,makanan,jumlah)
```
