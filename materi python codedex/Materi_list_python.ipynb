1. grocery

```python
belanja = ["tomatoes", "potatoes", "onions", "carrots"]
print(f'list belanja: {belanja}')

#menambahkan "spinach" ke daftar belanja
belanja.append("spinach")
print(f'list belanja setelah ditambah spinach: {belanja}')

#Ubah item pertama dalam list menjadi "cabbage"
belanja[0] = 'cabbage'
print(f'list belanja setelah item pertama diganti cabbage: {belanja}')

#Hapus item "onions" dari list
belanja.remove('onions')
print(f'list belanja setelah onion dihapus: {belanja}')

#Cetak berapa jumlah item di dalam list sekarang.
print(f'jumlah item di list belanja saat ini: {len(belanja)}')

#mengurutkan list berdasarkan abjad
belanja.sort()
print(f'list belanja setelah diurutkan: {belanja}')
```

2. to-do

```python
print('>>>>>to-do part 1<<<<<')
kegiatan =["Study Python", "Do laundry", "Buy groceries", "Call mom"]
print(f'list kegiatan yang dilakukan hari ini {kegiatan}')

#Tambahkan "Go jogging" ke dalam list todo
kegiatan.append('Go Jogging')
print(f'list kegiatan setelah ditambah go jogging: {kegiatan}')

#Ubah item kedua ("Do laundry") menjadi "Do laundry tomorrow".
kegiatan[1] = 'Do laundry tomorrow'
print(f'list kegiatan setelah item 2 diganti: {kegiatan}')

#Hapus "Buy groceries" dari list todo.
kegiatan.remove('Buy groceries')
print(f'list kegiatan setelah buy groceries dihapus: {kegiatan}')

#Cetak jumlah item yang tersisa di todo.
print(f'jumlah kegiatan yang tersisa yaitu: {len(kegiatan)}')

#Cetak item terakhir di list todo menggunakan index negatif.
print(f'item terakhir di list kegiatan yaitu: {kegiatan[-1]}')

print('\n>>>>>to-do part 2<<<<<')
#Buat list todo dengan isi:
kegiatan = ["Write report","Practice coding","Water plants","Pay bills"]
print(f'kegiatan yang dilakukan: {kegiatan}')

#Tambahkan "Meditation" di posisi pertama list (index 0) dengan .insert().
kegiatan.insert(0, 'Meditation')
print(f'list kegiatan setelah  dimasukkan meditation pada index 0 dengan insert: {kegiatan}')

#Hapus tugas terakhir dari list dengan .pop().
kegiatan.pop()
print(f'list kegiatan setelah index terakhir dihapus dengan pop: {kegiatan}')

#Urutkan daftar todo secara alfabet dengan .sort().
kegiatan.sort()
print(f'list kegiatan setelah diurutkan sesuai abjad: {kegiatan}')

#Balik urutan list dengan .reverse().
kegiatan.reverse()
print(f'list kegiatan setelah urutan dibalik: {kegiatan}')

#Hapus tugas dari list dengan .pop(index).
kegiatan.pop(2)
print(f'list kegiatan setelah index 2 dihapus dengan .pop(): {kegiatan}')

#Cetak hasil akhir list todo dan jumlahnya.
print(f'list akhir kegiatan dan jumlahnya: {kegiatan}, jumlah: {len(kegiatan)}')
```

5. mixtape

```python
lagu = ["Blinding Lights","Levitating","Peaches","Stay"]
print(f'daftar lagu yang dimiliki: {lagu}')

#Tambahkan "Shape of You" di posisi pertama (index 0) dengan .insert().
lagu.insert(0,"Shape of You")
print(f'datar lagu setelah ditambah lagu shape of you di awal list: {lagu}')

#Tambahkan sekaligus beberapa lagu baru: "Bad Guy" dan "Senorita" dengan .extend().
lagu_tambahan =["Bad Guy","Senorita"]
lagu.extend(lagu_tambahan)
print(f'daftar lagu setelah ditambah lagu di list lagu_tambahan: {lagu}')

#Hapus lagu terakhir di mixtape dengan .pop().
lagu.pop()
print(f'daftar lagu setelah lagu terakhir dihapus dari list: {lagu}')

#Urutkan mixtape sesuai abjad (.sort()).
lagu.sort()
print(f'daftar lagu sesuai urutan abjad: {lagu}')

#Balik urutan mixtape (.reverse()).
lagu.reverse()
print(f'daftar lagu dengan urutan abjad terbalik: {lagu}')

#Cetak hasil akhir playlist dan jumlah lagunya.
print(f'daftar lagu: {lagu}, jumlah: {len(lagu)}')

#acak lagu untuk diputar
import random
random.shuffle(lagu)
print(f'daftar lagu diputar secara acak: {lagu}')
```

5. bucket list

```python
bucket_list =["Naik gunung","Belajar diving","Nonton konser","Keliling Jepang"]

#menampilkan daftar list berurutan
print('Daftar bucket list 2025:')
for i in range (len(bucket_list)):
  print(f'{i+1}. {bucket_list[i]}')

#menambahkan centang pada list yang terpenuhi
user = ''
while user != "exit":
  terpenuhi = input('silahkan masukkan nomor list yang sudah terpenuhi (beri koma bila lebih dari satu): ')
  selesai = terpenuhi.split(',')
  for i in selesai:
    bucket_list[(int(i))-1] += '✅'
  #Tampilkan kembali seluruh bucket list dengan status terbaru.
  print('daftar bucket list terbaru:')
  for i in range (len(bucket_list)):
    print(f'{i+1}. {bucket_list[i]}')
  user = input('masih ingin menambahkan centang untuk list yang selesai (ketik exit untuk keluar)? ').lower()


```
