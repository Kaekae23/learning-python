looping list

```python
buku = ['sang raja',
        'underling purpose',
        'sambal di ranjang',
        'jejak balak',
        'di tanah lada',
        'pegantin remaja']

print(f'jumlah buku = {len(buku)}')

for i in buku:
  print(i)
```

```python
list = ['ayam bakar','nasi padang','kepiting']
for i in list:
  print(i)

print()
for i in range(len(list)):
  print(list[i])

print()
for i in range(0,len(list)):
  print(list[i])

print()
i =0
a = len(list)
while i < a:
  print(list[i])
  i+=1
```
