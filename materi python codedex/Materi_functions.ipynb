3. Mars Orbiter (dengan return)

```python
#1. Buat function untuk menghitung waktu tempuh: waktu_tempuh(jarak, kecepatan)
def waktu_tempuh(jarak,kecepatan):
  t = jarak/kecepatan
  return t

#2. nilai pada fungsi yaitu Jarak ke Mars 225 juta km dan Kecepatan pesawat 75 ribukm/jam.
a = 225000000
b = 75000

#3. tampilkan hasil dalam jam dan hari
hasil = waktu_tempuh(a,b)
print(f'waktu tempuh pesawat ke mars: {hasil} jam')
print(f'hasil perhitungan dalam hari: {hasil/24} hari')

#4. Buat program agar user bisa input sendiri jarak dan kecepatan, hasilnya dalam hari
user1 = int(input('masukkan jarak tempuh (km): '))
user2 = int(input('masukkan kecepatan pesawat (km/jam): '))
hasil = waktu_tempuh(user1,user2)
print(f'waktu tempuh pesawat dengan jarak {user1} km dan kecepatan pesawat {user2} km/jam dalam yaitu: {hasil} jam')
print(f'waktu tempuh dalam hari: {hasil/24} hari')

#5.Buat function misi_mars(jarak, kecepatan, bahan_bakar_per_jam) yang:
#Menghitung waktu tempuh dalam jam.
#Menghitung total bahan bakar (liter) dengan cara waktu_tempuh_jam × bahan_bakar_per_jam.
#Mengembalikan kedua nilai itu (waktu tempuh dalam jam, total bahan bakar).
#Gunakan function ini untuk menghitung misi dengan data:
#Jarak = 225.000.000 km
#Kecepatan = 75.000 km/jam
#Bahan bakar per jam = 500 liter
def misi_mars(jarak,kecepatan,bahan_bakar):
  waktu = jarak/kecepatan
  energi = waktu*bahan_bakar
  return waktu, energi
a= 225000000
b= 75000
c= 500
lama, bensin = misi_mars(a,b,c)
print(f'waktu tempuh pesawat: {lama} jam atau {lama/24} hari')
print(f'bahan bakar yang diperlukan: {bensin:,.0f} liter')
```

4. Kalkulator Sederhana

```python
#Buatlah 4 function operasi dasar (+,-,/,*):
def tambah(a,b):
  return a+b
def kurang(a,b):
  return a-b
def bagi(a,b):
  return a/b
def kali(a,b):
  return a*b

#Buat menu pilihan (1 = Tambah, 2 = Kurang, dst)
menu = ["pertambahan", "pengurangan", "pembagian", "perkalian"]
for i in range(len(menu)):
  print(f'{i+1}. {menu[i]}')

#Program meminta user untuk memasukkan angka pertama dan kedua lalu jenis operasi
operasi = int(input('pilih jenis operasi perhitungan yang anda inginkan (1-4): '))
pilihan1 = int(input('masukkan angka pertama: '))
pilihan2 = int(input('masukkan angka kedua: '))

#Tampilkan hasil sesuai operasi yang dipilih user.
if (operasi>=1 and operasi<=4):
  if operasi==3 and pilihan2==0:
    print('Error! tidak bisa melakukan pembagian dengan 0')
  elif operasi==1:
    hasil = tambah(pilihan1,pilihan2)
    print(f'hasil perhitungan {menu[operasi-1]} dari angka {pilihan1} dan {pilihan2} yaitu: {hasil}')
  elif operasi==2:
    hasil = kurang(pilihan1,pilihan2)
    print(f'hasil perhitungan {menu[operasi-1]} dari angka {pilihan1} dan {pilihan2} yaitu: {hasil}')
  elif operasi==3:
    hasil = bagi(pilihan1,pilihan2)
    print(f'hasil perhitungan {menu[operasi-1]} dari angka {pilihan1} dan {pilihan2} yaitu: {hasil}')
  else:
    hasil = kali(pilihan1,pilihan2)
    print(f'hasil perhitungan {menu[operasi-1]} dari angka {pilihan1} dan {pilihan2} yaitu: {hasil}')
else:
  print('operasi yang anda pilih tidak tersedia')

```

```python
#buat program Python untuk menghitung keuntungan maksimal dari jual-beli saham, dengan aturan:
#Hanya boleh beli 1 kali dan jual 1 kali.
#Jual harus dilakukan setelah beli.
#Jika tidak ada peluang untung, hasilnya 0.
#Gunakan dua data harga saham berikut:

harga1 = [7, 1, 5, 3, 6, 4]
harga2 = [7, 6, 4, 3, 1]

#Buat fungsi bernama max_profit(harga) yang menerima list harga.
def max_profit(harga):
  min_harga = harga[0]
  max_untung = 0
  for i in harga:
    if i<=min_harga:
      min_harga = i
    else:
      untung = i - min_harga
      max_untung = max(max_untung,untung)
      print(f'harga beli {min_harga} dijual di {i} -> untung {untung}')
  if max_untung <=0:
    max_untung = "keuntungan lebih rendah dari harga beli"
  return max_untung

#Coba jalankan fungsi untuk harga1 dan harga2.
print(f'keuntungan maksimum penjualan saham 1: {max_profit(harga1)}')
print(f'keuntungan maksimum penjualan saham 2: {max_profit(harga2)}')
```

```python
#Buat function tampilkan_menu() yang menampilkan daftar menu

def tampilkan_menu(menu):
  print("Daftar Menu".center(30))
  for i in menu:
    print(f'{i[0]:<20} Rp.{i[1]:,}')

#Buat function hitung_total(pesanan) yang menghitung total harga.
def hitung_total(pesanan):
  harga=[]
  for i in pesanan:
    for j in menu:
      a = i.split(" ")
      if a[0].title() in j[0]:
        harga.append(j[1]*int(a[1]))
  total = sum(harga)
  return harga, total

#Buat function hitung_diskon(total) dengan aturan diskon 10% jika total ≥ Rp50.000
def hitung_diskon(total):
  if total>=50000:
    bayar = total*90/100
  else:
    bayar = total
  return bayar

#Buat function print_struk(pesanan, total, bayar) untuk menampilkan struk pembelian.
def print_struk(pesanan,total,bayar):
  print('Struk Pembelian'.center(30))
  for i in pesanan:
    for j in menu:
      a = i.split(" ")
      if a[0].title() in j[0]:
        print(f'{a[0]:<15} x {a[1]} Rp.{j[1]*int(a[1]):,}')
  print("-"*30)
  print(f'{"Total Harga:":<20} Rp.{total:,}')
  if total>=50000:
    print(f'{"diskon 10%:":<20} Rp.{(total-bayar):,}')
    print(f'{"bayar:":<20} Rp.{bayar:,}')


menu = [("Burger", 25000), ("Fried-Chicken", 20000), ("French-Fries", 15000), ("Cola", 10000), ("Ice-Cream", 8000)]

tampilkan_menu(menu)
user = input("tulis menu dan jumlah yang anda inginkan (pisahkan dengan koma): ")
pesanan = user.split(",")

harga, total = hitung_total(pesanan)
bayar = hitung_diskon(total)
print_struk(pesanan,total,bayar)
```
